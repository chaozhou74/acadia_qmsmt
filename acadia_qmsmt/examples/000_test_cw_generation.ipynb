# Generate continuous wave signal from a DAC channel

for testing DAC output and RF chain, etc

```python
from system_info import BOARD_IP
from acadia.runtimes.continuous_synthesis import ContinuousSynthesisRuntime
```

```python
rt = ContinuousSynthesisRuntime(channel="DAC0",  
                                amplitude = 0.5,  # (-1, 1) amplitude scale
                                mix_reconstruction=True, # if True, use mixed reconstruction mode (stronger signal in 2nd/3rd Nyquist zone), else use Zero-Order-Hold mode (stronger signal in 1st Nyquist zone)
                                frequency=8.7e9, # Hz
                                vop=2500, # gain of internal DAC amplifier, (2500-40000)
                                timeout=240 # seconds, will play this long unless stopped manually
                                )

rt.deploy(BOARD_IP, "acadia.runtimes.continuous_synthesis")

rt.display()

# was
```
